Analyzing top 5 directors

```python
df['director']=df['director'].fillna('Director not specified')
df.head()
```

```python
directors_list = pd.DataFrame()
print(directors_list)
```

```python
directors_list=df['director'].str.split(',',expand=True).stack()
print(directors_list)
```

```python
directors_list = directors_list.to_frame()
print(directors_list)
```

```python
directors_list.columns = ['Directors']
print(directors_list)
```

```python
directors = directors_list.groupby('Directors').size().reset_index(name='Total Count')
print(directors)
```

```python
directors = directors[directors.Directors != 'Director not specified']
print(directors)
```

```python
directors = directors.sort_values(by='Total Count', ascending=False)
print(directors)
```

```python
top5_directors= directors.head()
print(top5_directors)
```

```python
barChart = px.bar(top5_directors, x='Directors', y='Total Count', title='Top 5 Directors on Netflix')
barChart.show()
```

Analyzing top 5 Actors

```python
df['cast'] = df['cast'].fillna('Cast not specified')
df.head()
```

```python
cast_df = pd.DataFrame()
print(cast_df)
```

```python
cast_df = df['cast'].str.split(',', expand=True).stack()
print(cast_df)
```

```python
cast_df = cast_df.to_frame()
print(cast_df)
```

```python
cast_df.columns = ['Actors']
print(cast_df)
```

```python
actors = cast_df.groupby('Actors').size().reset_index(name='Total Count')
print(actors)
```

```python
actors = actors[actors.Actors != 'Cast not specified']
print(actors)
```

```python
actors = actors.sort_values(by='Total Count', ascending=False)
print(actors)
```

```python
actors.head()
```

```python
barChart2 = px.bar(actors.head(), x='Actors', y='Total Count', title='Top 5 Actors on Netflix')
barChart2.show()
```
